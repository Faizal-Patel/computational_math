## Exercise 1

Plot the Taylor approximation of $\frac{1}{1+x}$ of degree $3$ centered at $0$, for $x$ in $[0,5]$

````python
import sympy as sp
import numpy as np 
import matplotlib.pyplot as plt 


# define the functions
def F(x):
    return ((1)/(1+x))
def P3(x): 
    return (1 - x) + (x ** 2) - (x ** 3)
    

# pick a 100 points between 0 and 5 and plot
x = np.linspace(0,5,100)

plt.figure(1)
plt.clf()
plt.plot(x,F(x),label = "F(x)")
plt.plot(x,P3(x),label = "P3(x)")
plt.legend()
plt.show()
plt.show()
````

## Exercise 3

Find and plot on the $x-$interval $[0,2\pi]$: $\sin(x)$, $P_1(x)$, $P_2(x)$, $P_3(x), P_4(x)$, where the Taylor polynomials are centered at $\pi$.

````python
import numpy as np 
import matplotlib.pyplot as plt 
import math 

#attributes
pi = (np.pi)

# define the functions

def F(x):
    return np.sin(x)
    
def P1(x):
    return np.sin(pi) + (-1*(x - pi))
    
def P2(x):
    return np.sin(pi) + (-1*(x - pi))
    
def P3(x):   
    return np.sin(pi) + (-1*(x - pi)) + (1/6*((x - pi)**3))
     
def P4(x):   
    return np.sin(pi) + (-1*(x - pi)) + (1/6*((x - pi)**3))
     

#  pick a 100 points between 0 and 2(𝜋) and plot

x = np.linspace(0,2*pi,100)

plt.figure(1)
plt.clf()
plt.plot(x,F(x),label = "F(x) = sin(x)")
plt.plot(x,P1(x), label = "P1(x)")
plt.plot(x,P2(x),label = "P2(x)")
plt.plot(x,P3(x), label = "P3(x)")
plt.plot(x,P4(x),label = "P4(x)")
plt.legend()
plt.show()

````

## Exercise 4
1. Find and plot on the $x-$interval $[-1,1]$, the linear and quadratic Taylor polynomials of $e^{\tan(x)}$ centered at $x=0$.
2. Plot the error between the Taylor of polynomial of degree 2 for $e^{\tan(x)}$, centered at 0. 
3. Bound the error in this approximation when $-1\leq x \leq 1$. ($Error(x)=|Approximation(x) - Exact(x)|$)

````python
import sympy as sp
import numpy as np 
import matplotlib.pyplot as plt 


# define the functions


def F(x): 
    return np.exp(np.tan(x))
 
def P2(x): 
    return 1 + x + (x**2/2)
   
def Error(x): 
    return np.abs(P2(x) - F(x))


    
print("The error between P2(x) and F(x) at x = 0 is: ", str(Error(0))) 

print("\nThe bound in this approximation when −1 ≤ 𝑥 ≤ 1 seems to be: ", str(Error(1)))



# we pick a 100 points between -1 and 1
x = np.linspace(-1,1,100)


plt.figure(1)
plt.clf()
plt.plot(x,F(x),label = "F(x) = e^(tan(x))")
plt.plot(x,P2(x),label = "P2(x)")
plt.plot(x,Error(x),label = "Error(x)")
plt.legend()
plt.show()

````

## Exercise 5
Let $f(x)=\sqrt[3]{x}$
1. Find and plot $P_3(x)$ centered at $a=1$
2. Plot $P_3(x)$ and $f(x)$ for in $x\in[0,4]$
3. Plot the error between $P_3$ and $f$
3. Approxmation $\sqrt[3]{2}$ and $\sqrt[3]{3}$  using $P_3$. How far off are they?

````python
import sympy as sp
import numpy as np 
import matplotlib.pyplot as plt 


# define the functions

def F(x): 
    return (x**(1/3))
    
def P3(x):
    return (1) + ((x - 1) / 3) - (((x-1)**2)/ 9) + ((5 *((x - 1)**3)) / 81)
 
def Error(x):
    return np.abs(F(x) - P3(x))
      

print("The error between P3(x) and F(x) at x = 1 is: ", str(Error(1)))

print("\nF(x) approximated using x = 2 is: ", str(F(2)))
print("P3(x) approximated using x = 2 is: ", str(P3(2)))

print("\nF(x) approximated using x = 3 is: ", str(F(3)))
print("P3(x) approximated using x = 3 is:", str(P3(3)))

print("\nHow far off are they?")

print("\nThe error when x is 2 is:", str(Error(2)))
print("The error when x is 3 is:", str(Error(3)))


# we pick a 100 points between 0 and 4
x = np.linspace(0,4,100)
plt.figure(1)
plt.clf()
plt.plot(x,F(x),label = "F(x)")
plt.plot(x,P3(x),label = "P3(x)")
plt.plot(x,Error(x),label = "Error(x)")
plt.legend()
plt.show()
````

## Bonus

Use `scipy.interpolate.approximate_taylor_polynomial` to compute the Taylor approximation of degreee 1,2,3,4 of $\log(1+e^x)$ at centered at $0$ (plot them and the function)

````python
#https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.approximate_taylor_polynomial.htmlimport numpy as np

# referenced the above website in order to find the right syntax using 'approximate_taylor_polynomial' as well as label and plotting.

import numpy as np 
import matplotlib.pyplot as plt
from scipy.interpolate import approximate_taylor_polynomial

x = np.linspace(-10.0, 10.0, num=100)
plt.plot(x, np.log(1 + np.exp(x)), label="log(1+(e^x))")

for degree in np.arange(1, 5, step=1):
    log_taylor = approximate_taylor_polynomial(np.log, 2, degree, 1, order = degree + 1)
    plt.plot(x, log_taylor(x), label=f"degree={degree}")   
plt.legend()
plt.axis([-10, 10, -10, 10])
plt.show()
````
